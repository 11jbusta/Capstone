```python
def remove_outliers(df, column, lower_quantile=0.25, upper_quantile=0.75):
    # Calculate the lower and upper quantiles
    Q1 = df[column].quantile(lower_quantile)
    Q3 = df[column].quantile(upper_quantile)
    IQR = Q3 - Q1

    # Determine outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Remove outliers from the 'FH' column in df6
df6_filtered = remove_outliers(df6, 'FH')

# Save the filtered dataframe back to a Parquet file if needed
df6_filtered.to_parquet('filtered_df6.parquet', engine='pyarrow')

# Print the size of the original and filtered dataframes
print(f'Original data size: {len(df6)}')
print(f'Filtered data size: {len(df6_filtered)}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Assuming df6_filtered is already loaded and available in the environment

# Fit KMeans with 8 clusters
kmeans = KMeans(n_clusters=8, random_state=42)
kmeans.fit(df6_filtered)

# Get the cluster labels
labels = kmeans.labels_

# Apply PCA to reduce to 2 principal components
pca = PCA(n_components=2)
pca_reduced = pca.fit_transform(df6_filtered)

# Get the top 5 features contributing to PCA1 and PCA2
pca_components = pca.components_

# Identify the top 5 features for PCA1
pca1_feature_importance = np.abs(pca_components[0])
top5_pca1_features_indices = np.argsort(pca1_feature_importance)[-5:]
top5_pca1_features = df6_filtered.columns[top5_pca1_features_indices]

# Identify the top 5 features for PCA2
pca2_feature_importance = np.abs(pca_components[1])
top5_pca2_features_indices = np.argsort(pca2_feature_importance)[-5:]
top5_pca2_features = df6_filtered.columns[top5_pca2_features_indices]

# Combine the top features
top_features = np.unique(np.concatenate([top5_pca1_features, top5_pca2_features]))

# Add cluster labels to the original DataFrame
df6_filtered['Cluster'] = labels

# Calculate the average values of the top features for each cluster
average_values_per_cluster = df6_filtered.groupby('Cluster')[top_features].mean()

# Plot the average values for each top feature per cluster
for feature in top_features:
    plt.figure(figsize=(10, 6))
    average_values_per_cluster[feature].plot(kind='bar', color='skyblue')
    plt.title(f'Average {feature} per Cluster')
    plt.xlabel('Cluster')
    plt.ylabel(f'Average {feature}')
    plt.show()
```
